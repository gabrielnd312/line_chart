#**Importing packages and defining our dataset**
First, let's import the necessary packages and choose our dataset to work with. For visualization we can create our own dataset, but for didactic purposes we will work with this one below about fertility rate in the world.

```python
#Let's import the necessary packages.
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python

#We import the .csv file for demo and preview its first entries.
data = "https://raw.githubusercontent.com/gabrielnd312/line_chart/main/fertility_across_nations.csv"
df = pd.read_csv(data, index_col = "Country Name")

df.head()

```

```python
#dataset size
print(f"Our dataframe for this demo has {df.shape[0]} rows and {df.shape[1]} columns.")
```

For visualization purposes only, we will select data from **Brazil**. To do this, we first have to transform the entries into ["Country Name"] in our index, so that we can select just one country and extract the information for each year.

```python
df_brazil = df.loc['Brazil']
df_brazil.head()
```

#**Plotting the graph**

<left><img alt="line_chart2" width="60%" src="https://img.freepik.com/fotos-gratis/conceito-de-negocio-com-holografia-grafica_23-2149160934.jpg?w=1380&t=st=1686923342~exp=1686923942~hmac=1e1eb2552f77b9c18ecf1999f24a5bc85ceef842e8280df609f825c6b2f948ef"></left>

```python
plt.plot(df_brazil[1:]);
```

This is the first and easiest way to plot our graph. However, you can verify that the information is not clear, as it is not even possible to identify the years that were considered.

For datasets with little information or for simpler analyses, it may be enough.

In our case, we still have to work on this graph.

```python
plt.plot(df_brazil[1:])

plt.xticks(rotation=90)


plt.show()
```


Using the **"rotation"** we make the information contained in the x axis easier to identify and that way we can have a better visualization of the information. **But it's still not good enough.**

As our dataset has many years of information, the numbers overlap, and we can correct this as follows:

```python
plt.plot(df_brazil[1:])

plt.xticks(rotation=90)

#We select the amount of "labels" that we want to visualize in the graph.
n_labels = 15

ind_labels = np.linspace(1, len(df_brazil.index)-1, n_labels, dtype=int)

selected_labels = df_brazil.index[ind_labels]

plt.xticks(ind_labels, selected_labels);

plt.tight_layout()
```

That way we get a much **cleaner** and **friendlier** graph.

We just subtract some years written on the X-axis and keep the labels rotated.

Note: Although we do not see the numbers for all years, **our complete dataset was considered in the graph**, so by doing this we do not lose any type of information.

#**Adding title and legends to our chart**

<left><img alt="line_chart2" width="60%" src="https://images.unsplash.com/photo-1591696205602-2f950c417cb9?ixlib=rb-4.0.3&ixid=M3wxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8fA%3D%3D&auto=format&fit=crop&w=1170&q=80"></left>


The graphic is mounted and visible, however, even if the viewer has the context behind it, it would be a little **difficult to understand** what the graphic is about just by looking at it.

The objective is that whoever analyzes your chart can understand it in the easiest possible way. For this, it is essential to put **necessary and relevant information** in each of its aspects.

For this we will use the methods:

**.title()**

**.xlabel()**

**.ylabel()**

In this way we can define:

*.title()* -> **Title of our graph, what it talks about, what is its scenario.**

*.xlabel()* -> **What is the information contained in the x-axis?**

*.ylabel()* -> **What is the information contained in the y-axis?**

This way reading and understanding the chart will be much easier and it will become more visually attractive.

```python
plt.plot(df_brazil[1:])

plt.title("Fertility Rates - Brazil")
plt.xlabel("Year")
plt.ylabel("Fertility Rate")
plt.xticks(rotation=90)

#We select the amount of "labels" that we want to visualize in the graph.
n_labels = 15

ind_labels = np.linspace(1, len(df_brazil.index)-1, n_labels, dtype=int)

selected_labels = df_brazil.index[ind_labels]

plt.xticks(ind_labels, selected_labels);

plt.tight_layout()
```

Of course, these are just superficial methods of treating graphics, however, the difference between **before and after** this cleaning is clear.

Following this line of work, you will be able to present **cleaner graphics** that pass a clearer message just by looking at it, fulfilling its role as an **information tool**.

Matplotlib also allows for a number of graph **modifications and customizations.**

Some examples will be placed below, but it is recommended that you look for the **library's official documentation** to verify several other customization and creation possibilities. You can access it [here](https://matplotlib.org/2.0.2/gallery.html).

```python
#To customize the line color, we use the "color" argument.
plt.figure(figsize = (5,5))
plt.plot(df_brazil[1:], color = "red")

plt.title("Fertility Rates - Brazil")
plt.xlabel("Year")
plt.ylabel("Fertility Rate")
plt.xticks(rotation=90)


n_labels = 15

ind_labels = np.linspace(1, len(df_brazil.index)-1, n_labels, dtype=int)

selected_labels = df_brazil.index[ind_labels]

plt.xticks(ind_labels, selected_labels);

plt.tight_layout()
```

It is possible to change the color by writing the **name of the color** as in the example above, which we use the color **("red")**, we can also use **hex code**, to define an exact color and standardize it.

In addition to the color, the **"figsize"** method was used to define the size of the graph. Depending on the presentation, a **larger** or **smaller** graphic will be important.

In this case, for better visualization purposes, we will use a chart smaller than the default size.

```python
#Changing color with hex code
#To customize the line color, we use the "color" argument.
plt.figure(figsize = (5,5))
plt.plot(df_brazil[1:], color = "#FF0000")

plt.title("Fertility Rates - Brazil")
plt.xlabel("Year")
plt.ylabel("Fertility Rate")
plt.xticks(rotation=90)


n_labels = 15

ind_labels = np.linspace(1, len(df_brazil.index)-1, n_labels, dtype=int)

selected_labels = df_brazil.index[ind_labels]

plt.xticks(ind_labels, selected_labels);

plt.tight_layout()
```

Here we have a replication that is exactly the same visually, however, we use the **red hex code** to modify the graph. We use the code **##FF0000** for the color.

#**Line Style**

We can also change the **graph line style.**

For example, we can use the arguments **"dotted"**, **"dashed"** and "**dashdot"**.

There is also the **"solid"** argument which is the default.

Check the results in the next lines.

```python
#Here we change the color to #117655 and the line style to "dotted"
plt.figure(figsize = (5,5))
plt.plot(df_brazil[1:], color = "#117655", linestyle = "dotted")

plt.title("Fertility Rates - Brazil")
plt.xlabel("Year")
plt.ylabel("Fertility Rate")
plt.xticks(rotation=90)


n_labels = 15

ind_labels = np.linspace(1, len(df_brazil.index)-1, n_labels, dtype=int)

selected_labels = df_brazil.index[ind_labels]

plt.xticks(ind_labels, selected_labels);

plt.tight_layout()
```

```python
#Here we change the color to #0e1e52 and the line style to "dashed"
plt.figure(figsize = (5,5))
plt.plot(df_brazil[1:], color = "#0e1e52", linestyle = "dashed")

plt.title("Fertility Rates - Brazil")
plt.xlabel("Year")
plt.ylabel("Fertility Rate")
plt.xticks(rotation=90)


n_labels = 15

ind_labels = np.linspace(1, len(df_brazil.index)-1, n_labels, dtype=int)

selected_labels = df_brazil.index[ind_labels]

plt.xticks(ind_labels, selected_labels);

plt.tight_layout()
```

```python
#Here we change the color to #f8f833 and the line style to "dashdot"
plt.figure(figsize = (5,5))
plt.plot(df_brazil[1:], color = "#f8f833", linestyle = "dashdot")

plt.title("Fertility Rates - Brazil")
plt.xlabel("Year")
plt.ylabel("Fertility Rate")
plt.xticks(rotation=90)

n_labels = 15

ind_labels = np.linspace(1, len(df_brazil.index)-1, n_labels, dtype=int)

selected_labels = df_brazil.index[ind_labels]

plt.xticks(ind_labels, selected_labels);

plt.tight_layout()
```

We use full forms to describe customization actions, but there are simpler options to make it easier:

"solid"(default) **->**	" - "

"dotted" **->** " : "

"dashed" **->** " -- "

"dashdot" **->** " -. "

For example, just replace **"dashdot"** with **"-."** in the code to customize your graph that way.


#**Plotting two or more graphs neatly**
It is common when working with graphics to have the need to create more than one for different categories, for different variables or in different contexts/dates.

If you don't pay attention, you can create graphs separately and leave your study either too long or disorganized, making it difficult for the reader to understand.

For this, you can use the following arguments in order to make the visualization clear:

```python
#Graphic 1
plt.subplot(1, 2, 1)  # 1 row, 2 columns, subplot 1

plt.plot(df_brazil[1:], color="#fe1e52", linestyle="dashed")
plt.title("Fertility Rates - Brazil")
plt.xlabel("Year")
plt.ylabel("Fertility Rate")
plt.xticks(rotation=90)

n_labels = 15
ind_labels = np.linspace(1, len(df_brazil.index) - 1, n_labels, dtype=int)
selected_labels = df_brazil.index[ind_labels]
plt.xticks(ind_labels, selected_labels)


#Graphic 2
plt.subplot(1, 2, 2)  # 1 row, 2 columns, subplot 2

plt.plot(df_brazil[1:], color="#ff8043", linestyle="dashdot")
plt.title("Fertility Rates - Brazil")
plt.xlabel("Year")
plt.ylabel("Fertility Rate")
plt.xticks(rotation=90)

n_labels = 6
ind_labels = np.linspace(1, len(df_brazil.index) - 1, n_labels, dtype=int)
selected_labels = df_brazil.index[ind_labels]
plt.xticks(ind_labels, selected_labels)


plt.tight_layout()
plt.show()
```

In this example, we use the **.subplot** function to create, within the same code, two different graphs.

Inside the parentheses we indicate (*#number_of_lines, #number of columns, #subplot*)

We can also do it this way:

```python
#Graphic 1
plt.subplot(2, 1, 1)  # 2 row, 1 columns, subplot 1

plt.plot(df_brazil[1:], color="#fe1e52", linestyle="dashed")
plt.title("Fertility Rates - Brazil")
plt.xlabel("Year")
plt.ylabel("Fertility Rate")
plt.xticks(rotation=90)

n_labels = 15
ind_labels = np.linspace(1, len(df_brazil.index) - 1, n_labels, dtype=int)
selected_labels = df_brazil.index[ind_labels]
plt.xticks(ind_labels, selected_labels)


#Graphic 2
plt.subplot(2, 1, 2)  # 2 row, 1 columns, subplot 2

plt.plot(df_brazil[1:], color="#ff8043", linestyle="dashdot")
plt.title("Fertility Rates - Brazil")
plt.xlabel("Year")
plt.ylabel("Fertility Rate")
plt.xticks(rotation=90)

n_labels = 6
ind_labels = np.linspace(1, len(df_brazil.index) - 1, n_labels, dtype=int)
selected_labels = df_brazil.index[ind_labels]
plt.xticks(ind_labels, selected_labels)


plt.tight_layout()
plt.show()
```


The same graphics but this time we consider only **1 column and 2 rows.**

Just define the way you want to present it inside the **.subplot()**

In many cases it is necessary to verify the information collected in a graph not in isolation, but by **comparing it with another reference.**

In this way, better than two graphs side by side, placing only one graph with two lines, each one representing each of the references, makes the visualization and **understanding of the comparison even easier.**

For demonstration purposes, we use the US fertility data in the **df_usa** DataFrame and place it on the same graph as the Brazil data.

```python
#Creating the dataframe
f_usa = df.loc['United States']

plt.plot(df_brazil[1:], color="#117650", linestyle="solid", label="Brazil")
plt.plot(df_usa[1:], color="#0e1e52", linestyle="solid", label="United States")

plt.title("Fertility Rates - Brazil")
plt.xlabel("Year")
plt.ylabel("Fertility Rate")
plt.xticks(rotation=90)

n_labels = 15
ind_labels = np.linspace(1, len(df_brazil.index) - 1, n_labels, dtype=int)
selected_labels = df_brazil.index[ind_labels]
plt.xticks(ind_labels, selected_labels)

plt.legend()

plt.tight_layout()
plt.show()
```

#**Line Width**

Another common example of customization, we will see that it is also possible to change the **line width** in the graph which can be useful for some presentations.

<left><img alt="line_chart2" width="50%" src="https://img.freepik.com/fotos-gratis/cerda-de-escova-e-esfregaco-amarelo_23-2147895492.jpg?w=360&t=st=1686923595~exp=1686924195~hmac=ce120d42386e5052377061a094172ff948a3ac06c994c7f8ff46c5f72f2ecb9e"></left>


For this we use the argument **"linewidth"** or just **"lw"**.

```python
#Chart using linewidth to change line thickness
plt.plot(df_brazil[1:], color="yellow", linewidth= 15)
plt.title("Fertility Rates - Brazil")
plt.xlabel("Year")
plt.ylabel("Fertility Rate")
plt.xticks(rotation=90)

n_labels = 15
ind_labels = np.linspace(1, len(df_brazil.index) - 1, n_labels, dtype=int)
selected_labels = df_brazil.index[ind_labels]
plt.xticks(ind_labels, selected_labels)

plt.tight_layout();
```

#**Transparency**

And as a last example, we will see below how to change the **transparency** of a line in a graph.

In some cases, when placing two lines on the same graph, you may want to highlight one between them and this transparency change can prove to be very useful to improve visualization.

For this, we use the **"alpha"** argument.

The **default is alpha = 1**, so if you want to make your graph line less visible, you can decrease this value. In the example, we use **alpha = 0.2**

```python
#Graphic 1 - alpha = 1
plt.subplot(1, 2, 1)  #

plt.plot(df_usa[1:], color="#000000", linestyle="dashdot", alpha=1)
plt.title("Fertility Rates - USA")
plt.xlabel("Year")
plt.ylabel("Fertility Rate")
plt.xticks(rotation=90)

n_labels = 6
ind_labels = np.linspace(1, len(df_brazil.index) - 1, n_labels, dtype=int)
selected_labels = df_brazil.index[ind_labels]
plt.xticks(ind_labels, selected_labels)


#Graphic 2 - alpha = 0.2
plt.subplot(1, 2, 2)

plt.plot(df_usa[1:], color="000000", linestyle="dashdot", alpha = 0.2)
plt.title("Fertility Rates - USA")
plt.xlabel("Year")
plt.ylabel("Fertility Rate")
plt.xticks(rotation=90)

n_labels = 6
ind_labels = np.linspace(1, len(df_brazil.index) - 1, n_labels, dtype=int)
selected_labels = df_brazil.index[ind_labels]
plt.xticks(ind_labels, selected_labels)


plt.tight_layout()
plt.show()
```
